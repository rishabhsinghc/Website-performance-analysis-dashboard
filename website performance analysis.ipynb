```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data-export (1).csv")
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df = df.drop(index = 0).reset_index(drop = True)
df.columns =["Channel group", "DateHours", "Users", "Session", "Engaged Session", "Average engagement time per session", "Engaged sessions per user	", "Events per session", "Engagement rate",	"Event count"]
```

```python
df.head()
```

```python
df["DateHours"] = pd.to_datetime(df["DateHours"], format="%Y%m%d%H", errors= 'coerce') 
df.head()
```

```python
numeric_cols = df. columns.drop(["Channel group", "DateHours"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors= 'coerce')
df["Hours"] = df["DateHours"].dt.hour
```

```python
df.head()
```

```python
df.describe()
```

# sessions and user over time

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
df.groupby("DateHours")[["Session", "Users"]].sum().plot(ax=plt.gca())
plt.title("sessions and users over time")
plt.xlabel("DateHours")
plt.ylabel("count")
plt.show()
```

# total users by channel

```python
plt.figure(figsize =(8,5))
sns.barplot(data=df, x="Channel group", y="Users", estimator=np.sum, palette="viridis")
plt.title("Total Users by channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

# Average engagement timeby channel

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Channel group", y="Average engagement time per session", estimator=np.mean, palette="magma")
plt.title("Avg engagement time by channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

# Engagement Rate Distribution by Channel

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Channel group", y="Engagement rate", palette="coolwarm")
plt.title("Engagement rate Distribition by Channel")
plt.xticks(rotation=45)
plt.show()
```

# Engaged vs Non Engaged Sessions

```python
session_df = df.groupby("Channel group")[['Session','Engaged Session']].sum().reset_index()
session_df["Non-Engaged"] = session_df["Session"] - session_df["Engaged Session"]

session_df_melted = session_df.melt(
    id_vars="Channel group", 
    value_vars=["Engaged Session", "Non-Engaged"], 
    var_name="Variable", 
    value_name="Value"
)

plt.figure(figsize=(8,5))
sns.barplot(data=session_df_melted, x="Channel group", y="Value", hue="Variable")
plt.title("Engaged vs Non Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

# trafic by hours and channel

```python
heatmap_data = df.groupby(["Hours", "Channel group"])["Session"].sum().unstack().fillna(0)

plt.figure(figsize=(8,5))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
plt.title("Traffic by hour and channel")
plt.ylabel("Hour of day")
plt.show()
```

# Engagement rate vs sessions over time

```python
df_plot = df.groupby("DateHours")[["Engagement rate", "Session"]].mean().reset_index()

plt.figure(figsize=(10,6))
plt.plot(df_plot["DateHours"], df_plot["Engagement rate"], label="Engagement rate", color="green")
plt.plot(df_plot["DateHours"], df_plot["Session"], label="Session", color="blue")
plt.title("Engagement rate vs session over time")
plt.legend()
plt.grid(True)
plt.show()
        
```
